```
import matplotlib.pyplot as plt
import pandas as pd

url = 'https://raw.githubusercontent.com/igorwiese/newcomerLabel/master/output.csv'

df = pd.read_csv(url,header=None)
df.columns = ['dontKnow', 'owner','repo','programming_language','stars','forks','total_issues']
df
```

```
#summary

#projects by language
print(df['programming_language'].value_counts())

```

```
#stars summary
print("Stars Median: ", df['stars'].median())
print("Stars Max: ",df['stars'].max())
print("Stars Min: ",df['stars'].min())
print("Stars percentil 90%: ",df.stars.quantile(0.9).round(1))

print("")

#forked summary
print("Forks Median: ",df['forks'].median())
print("Forks Max: ",df['forks'].max())
print("Forks Min: ",df['forks'].min())
print("Forks percentil 90%: ",df.forks.quantile(0.9).round(1))

print("")

#Total #Issues summary
print("Issues Median: ",df['total_issues'].median())
print("Issues Max: ",df['total_issues'].max())
print("Issues Min: ",df['total_issues'].min())
print("Issues percentil 90%: ",df.total_issues.quantile(0.9).round(1))
```

```
#we need to decide the number of samples we want and define the criterias
df_sample = df[(df.stars >= 436) & (df.forks >= 154) & (df.total_issues >= 491)].sample(200)

df_sample
```
